Peforming a K-Means fit of a 3-cluster data file that we created with drawscatter

```python
import pandas as pd
myDF = pd.read_csv("PyPals_data.csv")
kmeans = KMeans(n_clusters = 3)
kmeans.fit(myDF[['x', 'y']])

plt.scatter(myDF['x'], myDF['y'], c=kmeans.labels_)
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, color = 'k', edgecolors='y', linewidth=3)
plt.title('Three clusters')
plt.show()
```

#### We could also run a supervised learning classifier since we have the classes in the dataset

```python
DataFrame= pd.read_csv("PyPals_data.csv")
sns.scatterplot(x='x', y='y', hue='z', data= DataFrame)
plt.show()
```

Box Plot

```python
DataFrame= pd.read_csv("PyPals_data.csv")
sns.boxplot(x='z', y='x', data=DataFrame)
plt.title('Average x by Category')
plt.show()

sns.boxplot(x='z', y='y', data=DataFrame)
plt.title('Average y by Category')
plt.show()
```

Bar Chart

```python
DataFrame= pd.read_csv("PyPals_data.csv")
sum_data = DataFrame.groupby('z').sum()


sum_data.plot(kind='bar')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Sum of x and y by Category')


plt.show()
```

HeatMap

```python
DataFrame= pd.read_csv("PyPals_data.csv")
sns.heatmap(DataFrame.pivot_table(values='x', index='z', columns='y'))


plt.title('Heatmap of x and y values')


plt.show()
```
