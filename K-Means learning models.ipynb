```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("Data Detectives to PyPals_Data.csv")

# Calculate WCSS for different numbers of clusters
wcss_values = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    wcss_values.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss_values)
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

data = pd.read_csv("Data Detectives to PyPals_Data.csv")

# Calculated silhouette score for different numbers of clusters
silhouette_values = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    silhouette_values.append(silhouette_score(data, kmeans.labels_))

print("Silhouette Values:")
for i in range(len(silhouette_values)):
    print("Number of Clusters =", i+2, ": Silhouette Score =", silhouette_values[i])
    
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("Data Detectives to PyPals_Data.csv")

# Elbow plot to determine optimal number of clusters
wcss_values = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    wcss_values.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss_values)
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Silhouette plot to determine optimal number of clusters
silhouette_values = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    silhouette_values.append(silhouette_score(data, kmeans.labels_))
plt.plot(range(2, 11), silhouette_values)
plt.title('Silhouette Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Train KMeans clustering models
kmeans_elbow = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans_silhouette = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Fit models to data
kmeans_elbow.fit(data)
kmeans_silhouette.fit(data)

# Predict cluster assignments for data points
labels_elbow = kmeans_elbow.predict(data)
labels_silhouette = kmeans_silhouette.predict(data)
```
